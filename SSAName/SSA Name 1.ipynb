### Extract the entire lines with the conditions year > 2010 and the first 5 lines in every gender and year.

```python

import os
mypath = "/Users/kwool/C/midterm/namebystate";
with  open("/Users/kwool/C/midterm/allstate.txt", "w") as fwrite:
        for file in os.listdir("/Users/kwool/C/midterm/namebystate"):
            if file.endswith("TXT"):
                fname = os.path.join(mypath, file);
                # for all files which ends with TXT
                with open(fname) as fread:
                    prevyear = '1000';
                    for line in fread:
                        splitline = line.split(',');
                        if (prevyear != splitline[2]):
                            prevyear = splitline[2];
                            tupleFcnt = 0; tupleMcnt=0;
                            
                        if (splitline[2] > '2010'):
                            if  (tupleFcnt < 5 ) and (splitline[1] == 'F'):
                                tupleFcnt += 1;
                                #print (splitline);
                                fwrite.write(line);
                            elif (tupleMcnt < 5 ) and (splitline[1] == 'M'):
                                tupleMcnt += 1;
                                #print (splitline);
                                fwrite.write(line);
                            
                        
                            

            
```
